```python
df = pd.read_csv('시계열분석_전처리.csv', encoding='cp949')
df1 = pd.read_csv('시계열_전처리_20-21.csv', encoding='utf-8')
df2 = pd.read_csv('시계열_전처리_22.csv', encoding='utf-8')
df.head()
```

```python
# '범죄건수' 문자열 삭제
df['Date'] = df['Date'].str.replace('_범죄건수', '')

df['Date'] = df['Date'].str.replace(r'(\d+)년(\d+)월', r'20\1\2', regex=True)

df['Date'] = pd.to_datetime(df['Date'], format='%Y%m').dt.strftime('%Y%m')

df.head()
```

```python
# '범죄건수' 문자열 삭제
df1['Date'] = df1['Date'].str.replace('_범죄건수', '')

df1['Date'] = df1['Date'].str.replace(r'(\d+)년(\d+)월', r'20\1\2', regex=True)

df1['Date'] = pd.to_datetime(df1['Date'], format='%Y%m').dt.strftime('%Y%m')
```

```python
# '범죄건수' 문자열 삭제
df2['Date'] = df2['Date'].str.replace('_범죄건수', '')

df2['Date'] = df2['Date'].str.replace(r'(\d+)년(\d+)월', r'20\1\2', regex=True)

df2['Date'] = pd.to_datetime(df2['Date'], format='%Y%m').dt.strftime('%Y%m')
```

```python
# 정확도 검증

gu_name = ['강남구', 'Date']

last_2year = list()
for i in range(2022, 2024):
    for j in range(1, 13):
        last_2year.append(['%04d-%02d' % (i,j)])
        
        
last_2year = pd.DataFrame(last_2year, columns = ['ds'])
last_2year['ds']= pd.to_datetime(last_2year['ds'])

for i in range(len(df1['Date'])):
    date_time_str = df1['Date'][i]
    
    date = datetime.datetime.strptime(str(date_time_str), '%Y%m')  #datetime.datetime.strptime(date, "%Y-%m-%d")
    
    df1['Date'][i] = date

gangnam_predict = pd.DataFrame()

for i,name in enumerate(gu_name):
    data = pd.DataFrame()
    data = df1[[name,'Date']]
    data.columns = ['y','ds']
    data['ds'] = pd.to_datetime(data['ds'])

    model = Prophet().fit(data)
    print(i,"model_fit")
    forecast = model.predict(last_2year)
    print(i,"model_predict")
    gangnam_predict = pd.concat([gangnam_predict, forecast['yhat']], ignore_index = True, axis=1)
```

```python
gangnam_predict = pd.concat([gangnam_predict, forecast['ds']], ignore_index = True, axis=1)

gangnam_predict.columns = ['강남구', 'Date']

gangnam_predict.to_csv('gangnam_predict_20-21.csv', encoding='cp949',index = False)
```

```python
correct_predictions = sum(a == p for a, p in zip(df2['강남구'], predict_22['강남구']))
accuracy = correct_predictions / len(df2['강남구']) * 100

print(f'정확도: {accuracy:.2f}%')
```

```python
def calculate_mse(actual, predicted):
    squared_errors = [(a - p) ** 2 for a, p in zip(actual, predicted)]
    mse = sum(squared_errors) / len(actual)
    return mse

# MSE 계산
mse = calculate_mse(df2['강남구'], predict_22['강남구'])

print(f'MSE: {mse:.2f}')
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import cohen_kappa_score
from scipy.stats import spearmanr

predict_22['강남구'] = predict_22['강남구'].astype(float)
df2['강남구'] = df2['강남구'].astype(float)

accuracy_df = accuracy_score(df2['강남구'], predict_22['강남구'])
kappa_df = cohen_kappa_score(df2['강남구'], predict_22['강남구'])
spearman_corr_df = spearmanr(df2['강남구'], predict_22['강남구'])

print('정확도:', accuracy_df)
print('코헨의 카파:', kappa_df)
print('스피어만 순위 상관계수:', spearman_corr_df)
```
