```python
import pandas as pd
import os 
import numpy as np


# 합칠 CSV 파일들이 있는 디렉토리 경로
folder_path = 'C:/Users/경남교육청/Desktop/김민제/capston design/web/Time Series'

file_list = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

dataframes = []
for file in file_list:
    file_path = os.path.join(folder_path, file)
    # 11번째 행만 읽어오기
    df = pd.read_csv(file_path, encoding='utf-8', header=12, nrows=1)
    dataframes.append(df)

# 데이터프레임을 병합하고 하나의 CSV 파일로 저장
merged_data = pd.concat(dataframes)
merged_data.to_csv('one.csv', encoding='utf-8', index=False)
```
