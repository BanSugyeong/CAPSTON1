```python
import pandas as pd

rain = pd.read_csv("강수량분석자료_20202022.csv", encoding='cp949', )
temp = pd.read_csv('기온분석자료_20202022.csv', encoding='cp949')
humidity = pd.read_csv('습도분석자료_20202022.csv', encoding='cp949')
wind = pd.read_csv('풍속분석자료_20202022.csv', encoding='cp949')
```

```python
# 데이터에서 필요한 칼럼만 가져오기
new_columns = ['일시', '일최다강수량(mm)']
rain = rain[new_columns]

t_new_columns = ['평균최고기온(℃)', '평균최저기온(℃)']
temp = temp[t_new_columns]
wind = wind['최대풍속(m/s)']
humidity = humidity['최저습도(%rh)']


# 행 기준으로 데이터프레임 합치기
df = pd.concat([rain, temp, wind, humidity], axis=1)
df
```

```python
# NaN 값 삭제
df = df.dropna()

# 데이터 저장
df.to_csv('날씨데이터총합_변경.csv', encoding='cp949', index=False)
```
