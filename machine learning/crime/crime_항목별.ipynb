```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

%matplotlib inline
```

```python
df = pd.read_csv('범죄데이터_범죄율_총합.csv', encoding='cp949')
df.head()
```

```python
selected_columns = df.filter(like='살인')

df_murder = pd.concat([df['자치구'], selected_columns], axis=1)

df_murder.to_csv('crime_murder.csv', encoding='cp949', index=False)
```
